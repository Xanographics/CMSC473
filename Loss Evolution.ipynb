```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_acc  = np.load('loss_out/Train_Acc_Loss.npy')
train_comp = np.load('loss_out/Train_Comp_Loss.npy')
```

```python
val_acc  = np.load('loss_out/Val_Acc_Loss.npy')
val_comp = np.load('loss_out/Val_Comp_Loss.npy')
```

# Comparing Validation and Training Accuracy Losses

```python
plt.figure(figsize=(11,11))
plt.plot(train_acc, label='Training Accuracy Loss')
plt.plot(val_acc,   label='Validation Accuracy Loss')
plt.legend(bbox_to_anchor=(1.05,1))
plt.show()
```

# Comparing Validation and Training Compression Loss

```python
plt.figure(figsize=(11,11))
plt.plot(train_comp, label='Training Compression Loss')
plt.plot(val_comp,   label='Validation Compression Loss')
plt.legend(bbox_to_anchor=(1.05,1))
plt.show()
```

# Comparing Validation Accuracy & Compression Loss

```python
plt.figure(figsize=(11,11))
x_ticks = list(range(500))
plt.plot(x_ticks, val_comp, label='Validation Compression Loss')
plt.plot(x_ticks, val_acc,  label='Validation Accuracy Loss')
plt.legend(bbox_to_anchor=(1.05,1))
plt.show()
```

# Comparing Training Validation and Compression Loss

```python
plt.figure(figsize=(11,11))
x_ticks = list(range(500))
plt.plot(x_ticks, train_comp, label='Training Compression Loss')
plt.plot(x_ticks, train_acc,  label='Training Accuracy Loss')
plt.legend(bbox_to_anchor=(1.05,1))
plt.show()
```
